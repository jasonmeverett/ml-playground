```python
import torch
from d2l import torch as d2l
from jevml.models.linear import SimpleLinear
from jevml.data.simple import SyntheticRegressionData
from jevml.trainers.simple import SimpleTrainer
```

```python
model = SimpleLinear(2, lr=0.03, sigma=0)
data = SyntheticRegressionData(w=torch.tensor([2, -3.4]), b=4.2, noise=0.03, batch_size=32)
trainer = SimpleTrainer(max_epochs=3)
trainer.fit(model, data)
with torch.no_grad():
    print(f'error in estimating w: {data.w - model.w.reshape(data.w.shape)}')
    print(f'error in estimating b: {data.b - model.b}')
```
