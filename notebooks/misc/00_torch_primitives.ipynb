## MPS Optimizations

```python
d = 10000
Y = torch.randn([d,d])
Z = torch.randn([d,d])

t0 = time.time()
torch.mm(Y,Z)
t1 = time.time()
print(f"CPU Elapsed: {t1-t0}")

Y_mps = Y.to("mps")
Z_mps = Z.to("mps")
t0 = time.time()
torch.mm(Y_mps,Z_mps)
t1 = time.time()
print(f"MPS Elapsed: {t1-t0}")
```
