```python
%matplotlib inline
import time
import torch
import torchvision
from torchvision import transforms
from d2l import torch as d2l

from jevml.data.mnist import FashionMNIST
from jevml.models.softmax import SoftmaxRegression
from jevml.trainers.simple import SimpleTrainer

d2l.use_svg_display()
```

```python
# Load in the dataset.
data = FashionMNIST()
tup = next(iter(data.val_dataloader()))
data.visualize(tup)
```

```python
data = FashionMNIST(batch_size=256)
model = SoftmaxRegression(num_inputs=784, num_outputs=10, lr=0.1)
trainer = SimpleTrainer(max_epochs = 15)
trainer.fit(model, data)
```

```python
# Now let's take a look at the failed predictions.

# Grab a batch from validation
X, y = next(iter(data.val_dataloader()))
y_hat = model(X)
preds = y_hat.argmax(axis=1)

bad_cases = (preds.type(y.dtype) != y)
X, y, preds = X[bad_cases], y[bad_cases], preds[bad_cases]
labels = [a+'\n'+b for a,b in zip(
    data.text_labels(y),
    data.text_labels(preds)
)]
data.visualize(
    [X, y], labels=labels
)
```
